```python
from sklearn import linear_model
from sklearn.svm import SVC
import sklearn.model_selection
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#dropped FID and Prate as they d not have any imapct of covid impact level
df.drop(['FID'], axis = 1, inplace=True)
df.drop(['PRate'], axis = 1, inplace=True)

df.head(5)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
#data -> attributes, target -> COVIDImpactLevel
data = df.drop(['COVIDImpactLevel'], axis=1)
target = df[['COVIDImpactLevel']]

#seperate training set and test set
train_X, test_X,train_Y,test_Y = train_test_split(data,target,test_size=0.3)

#create linear regression obj
lr_regr = linear_model.LinearRegression()
print(lr_regr)
#training via linear regression model
lr_regr.fit(train_X, train_Y)

#make prediction using the test set
lr_pred = lr_regr.predict(test_X)
lr_score = lr_regr.score(test_X, test_Y)
lr_mse=mean_squared_error(test_Y, lr_pred)
print('LRr_Coefficients: ', lr_regr.coef_)
print('LR_Mean Square Error: %.2f' % mean_squared_error(test_Y, lr_pred))
print('LR_Variance score: %.2f' % r2_score(test_Y, lr_pred))
print('Score: %.2f' % lr_regr.score(test_X, test_Y))
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_score
#create adaboost classification obj
ab_clf = AdaBoostClassifier(base_estimator=DecisionTreeClassifier(), n_estimators=100, 
                            learning_rate=0.5, random_state=100)

#training via adaboost classficiation model
ab_clf.fit(train_X, train_Y)
print("training....\n")

#make prediction using the test set
ab_pred = ab_clf.predict(test_X)
print('prediction: \n', ab_pred)

print('\nparms: \n', ab_clf.get_params)

#score
ab_clf_score = ab_clf.score(test_X, test_Y)
print("\nScore: %.2f" % ab_clf.score(test_X, test_Y))
ab_clf_mse = mean_squared_error(test_Y, ab_pred)
print("\nMSE: %.2f" % ab_clf.score(test_X, test_Y))
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.neighbors import KNeighborsClassifier
#creating bagging classifier
bagging = BaggingClassifier(base_estimator= DecisionTreeClassifier(), max_samples = 0.5, max_features = 0.5, 
                            bootstrap = False, bootstrap_features = False)
#fitting data to the model created
bagging.fit(train_X, train_Y)
#predicting the class based on test data and model fit
bg_pred = bagging.predict(test_X)
#clauclating score of model
bg_dt_score = bagging.score(test_X, test_Y)
bagging.score(test_X, test_Y)
#calculating MSE of model
bg_dt_mse=mean_squared_error(test_Y, bg_pred)
```

```python
import numpy as np
#creating bagging classifier
bagging = BaggingClassifier(base_estimator= KNeighborsClassifier(), max_samples = 0.5, max_features = 0.5, 
                            bootstrap = False, bootstrap_features = False)
#fitting data to the model created
bagging.fit(train_X, train_Y)
#predicting the class based on test data and model fit
bg_pred = bagging.predict(test_X)
#clauclating score of model
bg_score = bagging.score(test_X, test_Y)
bagging.score(test_X, test_Y)
#calculating MSE of model
bg_mse=mean_squared_error(test_Y, bg_pred)

print(np.sqrt(bg_mse))
```
